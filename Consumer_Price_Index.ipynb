```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df=pd.read_csv('/content/All_India_Index_july2019_20Aug2020_dec20_1_2.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isna().sum()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
df1=df.loc[(df['Year']==2013)&(df['Sector']=='Rural')]
df1.head()
```

```python
y=df1.drop(['Year'],axis=1)

y.plot.bar(x='Month')
plt.xlabel('Months')
plt.ylabel('CPI')
plt.title('RURAL:-Consumer Price Index in 2013')
plt.legend(loc='upper left',bbox_to_anchor=(1,1))
plt.rcParams["figure.figsize"] = (20, 20)
plt.show()
```
