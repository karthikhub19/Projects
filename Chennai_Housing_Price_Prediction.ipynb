```python
x=df['N_BEDROOM'].mode()[0]
df['N_BEDROOM'].fillna(x,inplace=True)
```

```python
y=df['N_BATHROOM'].mode()[0]
df['N_BATHROOM'].fillna(y,inplace=True)
```

```python
df.dtypes
```

```python
df[['DATE_SALE','DATE_BUILD']]=df[['DATE_SALE','DATE_BUILD']].apply(pd.to_datetime)
```

```python
df[['N_BEDROOM','N_BATHROOM']]=df[['N_BEDROOM','N_BATHROOM']].astype(int)
```

```python
df.dtypes
```

```python
df['TOTAL_PRICE']=df['REG_FEE']+df['COMMIS']+df['SALES_PRICE']
```
